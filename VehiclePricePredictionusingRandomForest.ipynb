```python
#handling the missing value
#data encoding
#feature scaling


from sklearn.impute import SimpleImputer #for missing value
from sklearn.preprocessing import OneHotEncoder #for encoding
from sklearn.preprocessing import StandardScaler #for scaling


from sklearn.pipeline import Pipeline #a sequence of data transformer
from sklearn.compose import ColumnTransformer #Groups all the pipeline steps for each of the columns
```

```python
df.dtypes
```

```python
cat_cols=["make","model","engine","fuel","transmission","trim","body","exterior_color","interior_color","drivetrain"]
num_cols=["year","cylinders","mileage","doors"]
```

```python
#feature engineering automation using pipeline and columntransformer

num_pipeline = Pipeline(steps = [('imputation', SimpleImputer(strategy = "median")),
                                ('scaling', StandardScaler())])
cat_pipeline = Pipeline(steps=[
    ('imputation', SimpleImputer(strategy='most_frequent')),
    ('encoding', OneHotEncoder(handle_unknown='ignore'))
])
```

```python
preprocessor = ColumnTransformer([("num_pipeline", num_pipeline, num_cols),
                  ("cat_pipeline", cat_pipeline, cat_cols)])
```

```python
preprocessor
```

```python
# Make sure to use the original DataFrames for transformation
x_train = preprocessor.fit_transform(x_train)
x_test = preprocessor.transform(x_test)
```

```python
x_train
```

```python
#now data is ready, lets build the multiple models

from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
```

```python
models={"support vector Regressor":SVR(),
        "DT Regressor":DecisionTreeRegressor(),
        "Multiple Linear Regression":LinearRegression(),
        "Random Forest Regressor":RandomForestRegressor()}
```

```python
from sklearn.metrics import r2_score
def model_train_eval(X_train, y_train, X_test, y_test, models):
    evaluation = {}
    for i in range(len(models)):
        model = list(models.values())[i]
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        model_score = r2_score(y_test, y_pred)
        evaluation[list(models.keys())[i]] = model_score
    return evaluation
```

```python
model_train_eval(x_train,y_train,x_test,y_test,models)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfr=RandomForestRegressor(oob_score=True)
```

```python
rfr
```

```python
#Hyperparameter Tuning
from sklearn.model_selection import RandomizedSearchCV
params={'max_depth':[1,50, 100, 150,200],
        'n_estimators':[50, 100,200]}
```

```python
params
```

```python
reg=RandomizedSearchCV(rfr,param_distributions=params,cv=5,verbose=3,scoring='r2',n_iter=5)
```

```python
reg
```

```python
reg.fit(x_train,y_train)
```

```python
print(reg.best_params_)
print(reg.best_score_)
reg.best_estimator_
```

```python
best_model=reg.best_estimator_
y_pred=best_model.predict(x_test)
```
